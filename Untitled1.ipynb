```python
# https://drive.google.com/drive/folders/1VIzYHCqp-S_NfqVjO0TtfxBFZUPotHa2?usp=sharing
# https://drive.google.com/file/d/1vbj8xow94noy_DB0o5g5deraEmp8d179/view?usp=sharing


import scipy.io
mat_file = scipy.io.loadmat('/content/drive/MyDrive/imagelabels.mat')

if 'labels' in mat_file:
  labels_array = mat_file['labels']
  print(f"Shape of the labels array: {labels_array.shape}")
else:
  print("Variable 'labels' not found in the .mat file. Please check the variable name.")
```

```python
%%time
first = True
IMG_SIZE = 64



files_path = '/content/drive/MyDrive/102flowers/jpg'

def get_image_number(filename):
  """Extracts the image number from a filename like 'image_00001.jpg'."""
  parts = filename.split('_')
  if len(parts) > 1:
    try:
      # Extract the part after '_' and before '.jpg', then convert to int
      number_str = parts[-1].split('.')[0]
      return int(number_str)
    except ValueError:
      return float('inf') # Put non-standard filenames at the end
  return float('inf') # Put filenames without '_' at the end


items = sorted(glob.glob(files_path + '/*.*'), key=get_image_number)
print(items)
tmpx = np.array([(cv2.resize(cv2.cvtColor(cv2.imread(x, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB), (IMG_SIZE,IMG_SIZE))) for x in items])
tmpy = np.array(labels_array)
tmpy = tmpy.reshape(-1, 1)

print(tmpx.shape)
print(tmpy.shape)
```

```python
counts = np.bincount(tmpy.flatten())
print(sorted(counts))
selected_indices = np.where((counts >= 93) & (counts < 112))[0]


filtered_indices = np.isin(tmpy.flatten(), selected_indices)
tmpx_filtered = tmpx[filtered_indices]
tmpy_filtered = tmpy[filtered_indices]

print(f"Shape of filtered tmpx: {tmpx_filtered.shape}")
print(f"Shape of filtered tmpy: {tmpy_filtered.shape}")

print(np.unique(tmpy_filtered).tolist())

cls_map = {}
for i, idx in enumerate(np.unique(tmpy_filtered).tolist()):
  cls_map[idx] = i

print(cls_map)

tmpy_filtered = np.array([cls_map[label] for label in tmpy_filtered.flatten()]).reshape(-1, 1)
print(np.unique(tmpy_filtered))

print("\nShape after remapping:")
print(tmpx_filtered.shape)
print(tmpy_filtered.shape)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(tmpx_filtered, tmpy_filtered, test_size=0.1, random_state=123)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)

train_data = xtrain.copy()
train_labels = ytrain.copy()
test_data = xtest.copy()
test_labels = ytest.copy()
```

```python
train_data = train_data / 255.0
test_data = test_data / 255.0

print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
print(test_labels.shape)
print(test_labels[0])
print(train_data[0])
```

```python
# 3. Keras Model 정의, Model summary -> 3 Conv Layer + ReLU + Dropout으로 구성할 것 (출력 예 참조)
model = tf.keras.models.Sequential()
#input layer
model.add(tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3)))
# conv 1
model.add(tf.keras.layers.Conv2D(64, 3, padding='same'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Activation('relu'))
# conv 2
model.add(tf.keras.layers.Conv2D(96, 3, 2, padding='same'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Activation('relu'))
# conv 3
model.add(tf.keras.layers.Conv2D(96, 3, 2, padding='same'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Activation('relu'))
# dense layers
model.add(tf.keras.layers.Conv2D(128, 3, 2, padding='same'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Activation('relu'))

model.add(tf.keras.layers.Conv2D(128, 3, 2, padding='same'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Activation('relu'))

model.add(tf.keras.layers.Flatten(name='flatten'))
model.add(tf.keras.layers.Dense(128))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Activation('relu'))

model.add(tf.keras.layers.Dense(9, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# prompt: label 22개까지 5개씩 그리기 (총 110개)

def plot_images_by_label(data, labels, num_images_per_label=5, num_labels_to_plot=22):
    """
    Displays a few images for each of the first `num_labels_to_plot` labels.

    Args:
        data (np.ndarray): The image data.
        labels (np.ndarray): The corresponding labels for the images.
        num_images_per_label (int): The number of images to display per label.
        num_labels_to_plot (int): The maximum number of unique labels to plot.
    """
    unique_labels = np.unique(labels)
    labels_to_plot = unique_labels[:num_labels_to_plot]

    # Calculate grid size
    n_cols = num_images_per_label
    n_rows = min(len(labels_to_plot), num_labels_to_plot)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3, n_rows * 3))
    fig.suptitle("Sample Images per Label")

    axes = axes.flatten() # Flatten axes array for easy iteration

    img_count = 0
    for i, label in enumerate(labels_to_plot):
        if i >= num_labels_to_plot:
            break # Stop if we have plotted enough labels

        label_indices = np.where(labels == label)[0]
        selected_indices = label_indices[:num_images_per_label]

        for j, idx in enumerate(selected_indices):
            ax = axes[img_count]
            ax.imshow(data[idx])
            ax.set_title(f"Label: {label}")
            ax.axis('off')
            img_count += 1

    # Hide any unused subplots
    for i in range(img_count, len(axes)):
        fig.delaxes(axes[i])

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for suptitle
    plt.show()



# Example usage:
plot_images_by_label(test_data, test_labels.flatten(), num_images_per_label=5, num_labels_to_plot=9)
```

```python
# @title 기본 제목 텍스트
%%time
history = model.fit(train_data, train_labels, epochs=50, batch_size=8,
                 validation_data=(test_data, test_labels) )
```

```python
plt.figure(figsize=(6, 10))
plt.subplot(2, 1, 1)
plt.title('Accuracy')
plt.plot(history.history['accuracy'], 'b', label='train_accuracy')
plt.plot(history.history['val_accuracy'], 'g', label='val_accuracy')
plt.ylim([0.7,1])
plt.grid(True)
plt.ylabel('Accuracy')
plt.legend(loc='best')

plt.subplot(2, 1, 2)
plt.title('Loss')
plt.plot(history.history['loss'], 'b', label='train_loss')
plt.plot(history.history['val_loss'], 'g', label='val_loss')
plt.ylim([0,0.5])
plt.grid(True)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

```python

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_baseline_model = converter.convert()
litert_model_path = "/content"
import pathlib

tflite_models_dir = pathlib.Path(litert_model_path)
tflite_models_dir.mkdir(exist_ok=True, parents=True)

tflite_baseline_model_file = tflite_models_dir/"fm_qat_model_bl.tflite"
size_baseline_model = tflite_baseline_model_file.write_bytes(tflite_baseline_model)
interpreter_base = tf.lite.Interpreter(model_content=tflite_baseline_model)
interpreter_base.allocate_tensors()
q_aware_model = tfmot.quantization.keras.quantize_model(model)

q_aware_model.compile(optimizer='adam',
                      loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                      metrics=['accuracy'])

q_aware_model.summary()
```
